### Number Analyzer Tool
* **What it does**:
* Checks if number is:
* Prime / Composite
* Even / Odd
* Palindrome
* Armstrong
* **Shows all results together:**

* **Concepts used:**
* Loops → prime logic
* Conditionals → checks
* Functions → separate function per check

```python
def menu():
    while True:
        print("\n----- MENU IS -----")
        print("1] Prime / Composite Number")
        print("2] Even / Odd Number")
        print("3] Palindrome Number")
        print("4] Armstrong Number")
        print("5] Exit")
        print()

        choice = int(input("Enter your choice: "))

        if choice == 1:
            prime_composite()
        elif choice == 2:
            even_odd()
        elif choice == 3:
            palindrome()
        elif choice == 4:
            armstrong()
        elif choice == 5:
            print("Exiting the Analyzer...")
            break  
        else:
            print("Enter a valid choice!")

def prime_composite():
    num = int(input("Enter the number:"))

    if num <= 1:
        print("Neither prime nor composite")
        return

    count = 0
    for i in range(1, num + 1):
        if num % i == 0:
            count += 1

    if count == 2:
        print(f"{num} is prime")
        print(f"{num} is not composite")
    else:
        print(f"{num} is not prime")
        print(f"{num} is composite")


def even_odd():
    num = int(input("Enter the number:"))
    if num % 2 == 0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")

def palindrome():
    num = int(input("Enter the number:"))
    rev = 0
    temp = num
    while temp > 0:
        digit = temp % 10
        rev = rev * 10 + digit
        temp = temp // 10

    if rev == num:
        print(f"{num} is palindrome")
    else:
        print(f"{num} is not a palindrome")

def armstrong():
    num = int(input("Enter the number:"))
    org = num
    digits = 0
    temp = num
    sum_total = 0
    
    #count digits
    while temp > 0:
        digits = digits + 1
        temp = temp // 10

    temp = num
        
    #calculate armstrong sum
    while temp > 0:
        digit = temp % 10
        sum_total += digit ** digits
        temp = temp // 10
    
    if sum_total == org:
        print(f"{num} is armstrong")
    else:
        print(f"{num} is not armstrong")

def __main__():
    menu()

__main__()
```
