#### Banking Interest Calculator
* **What it does:**
* Calculates:
* Simple Interest
* Compound Interest
Based on user choice

* **Concepts used:**
* Functions → SI & CI separately
* Conditionals → choice
* Loop → repeat calculations

```python
def menu():
    while True:
        print("---------- Welcome to our banking services!!! ----------")
        print("Please choose the service you want to use:")
        print("1] calculate simple interest")
        print("2] calculate compound interest")
        print("3] Exit")
        choice = int(input("Enter your choice:"))
        if choice == 1:
            simple_interest()
            print()
        elif choice == 2:
            compound_interest()
            print()
        else:
            print("Exiting the program.......")
            break

def simple_interest():
    P = float(input("Enter the principal amount:"))
    R = float(input("Enter the interest rate:"))
    T = float(input("Enter the number of years:"))
    SI = P * R * T / 100
    Total_amount = P + SI
    print(f"The new total amount is: {Total_amount}")

def compound_interest():
    P1 = float(input("Enter the principal amount:"))
    R1 = float(input("Enter the interest rate:"))
    T1 = float(input("Enter the number of years:"))
    A = P1 * (1 + R1 / 100) ** T1
    CI = A - P1 
    print(f"Your new total amount is: {CI}")

def main():
    menu()

main()
```
